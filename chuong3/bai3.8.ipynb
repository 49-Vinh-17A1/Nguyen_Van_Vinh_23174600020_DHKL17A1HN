```python
import numpy as np 
```

## 1

```python
with open('C:\\code\\gitpythonnangcao\\baitap pỵthon\\chuong3\\heights.txt','r',encoding = 'utf-8-sig') as file:
    for i in file:
        data = i.split(',')
np_heights = np.array(data,dtype=float)
np_heights[:5]
```

```python
with open('C:\\code\\gitpythonnangcao\\baitap pỵthon\\chuong3\\positions.txt','r',encoding = 'utf-8-sig') as file:
    for i in file:
        data = i.replace('[','').replace(']','').replace("'",'').replace(' ','').split(',')
np_positions = np.array(data)
np_positions[:5]
```

# 4

```python
dtype = [('position','U5'),('height',float)]
player = np.array(list(zip(np_positions,np_heights)),dtype= dtype)
player[10:20]
```

# 2

```python
player_GK = player[player['position']=='GK']
player_GK
```

```python
height_GK = player_GK['height']
print('chiều cao trung bình GK là:',np.mean(height_GK))
```

# 3

```python
player_diffrent_GK = player[player['position'] != 'GK']
player_diffrent_GK[:5]
```

```python
height_diffrent_GK = player_diffrent_GK['height']
```

```python
print("chiều cao trung bình của những vị trí khác là: ",np.mean(height_diffrent_GK))
```

```python
player = np.sort(player, order='height')
```

```python
print("vị trí có chiều cao cao nhất là: ",player[-1])
print("vị trí có chiều cao thấp nhất là: ",player[0])
```
